# Measure of Central Tendancy

* mean
* median
* Mode

```python
import numpy as np
ages=[12,56,33,16,34,23,67,14,29,21,46,41,45,20,20,120]
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics as st
print(st.mean(ages))
print(st.median(ages))
print(st.mode(ages))
```

# How To find Outlier

```python
import seaborn as sns
sns.boxplot(ages)
```

# 5 Number Summary

```python
q1,q3=np.percentile(ages,[25,75])
print(q1)
print(q3)
```

```python
## to check outlier [lowerfence-higher fence]
IQR=q3-q1
print(IQR)
lower_fence=q1-1.5*(IQR)
higher_fence=q3+1.5*(IQR)
print(lower_fence,higher_fence)
```

# Measure Of Dispersion
* 1.Variance
* 2.Standard Deviation

```python
# samle variance
st.variance(ages)
```

```python
# Population variance
np.var(ages,axis=0)
```

```python
def variance(data):
    n=len(ages)
    mean=sum(ages)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/n
    return variance
```

```python
variance(ages)
```

```python
def variance(data):
    n=len(ages)
    mean=sum(ages)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-1)
    return variance
```

```python
variance(ages)
```

```python
st.pvariance(ages) # for population
```

```python
import math
math.sqrt(st.variance(ages))
```

# Check Wether Distribution is Normal distribution

```python
## Q-Q Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## Create a Normal Distributed Data
f=np.random.normal(mu,sigma,1000)
plot_data(f)
```

```python
plot_data(df['SepalWidthCm'])
```

```python
plot_data(df['SepalLengthCm'])
```

```python
plot_data(df['PetalLengthCm'])
```

```python
plot_data(df['PetalWidthCm'])
```
