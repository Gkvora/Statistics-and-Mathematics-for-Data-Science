# Measure of Dispersion
1.variance
2.Standard deviation

```python
st.variance(b) #statistics a sample variance aape which is in formula devided by n-1
```

```python
np.var(b,axis=0) # numpy is given population varince which is in formula devided by n
```

```python
def variance(data):
    n=len(data)
    mean=sum(data)/n
    std=[(x-mean)**2 for x in data]
    variance=sum(std)/n
    return variance
```

```python
variance(b)
```

```python
def variance(data):
    n=len(data)
    mean=sum(data)/n
    std=[(x-mean)**2 for x in data]
    variance=sum(std)/(n-1)
    return variance
```

```python
variance(b)
```

```python
def variance(data,dof):
    n=len(data)
    mean=sum(data)/n
    std=[(x-mean)**2 for x in data]
    variance=sum(std)/(n-dof)
    return variance
```

```python
variance(b,1)
```

# Check whether distribution is normal distribution

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['SepalWidthCm'])
```

```python
mu, sigma = 3., 1. # mean and standard deviation
sample = np.random.lognormal(mu, sigma, 100)
```

```python
def plot_data_log(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(np.log(sample),dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data_log(sample)
```
